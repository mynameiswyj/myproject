## Statistics > ANOVA

ANOVA(Analysis of Variance)는 그룹 간의 평균 차이를 비교하기 위해 사용됩니다. 주로 세 개 이상의 그룹 간의 평균 차이를 검정하는 데 사용되며, 그룹 간의 변동과 그룹 내의 변동을 비교하여 평균 차이를 평가합니다.

### One-way ANOVA (일원분산분석)

예제) Pclass에 따라 Survived에 영향을 미치는가를 분석합니다.

```
# Visual Python: ANOVA
from scipy import stats
def vp_confidence_interval(var, confidence_level=0.95):
    try: sr = pd.Series(var)
    except: return np.nan
    return stats.t.interval(confidence_level, df=sr.count()-1, loc=sr.mean(), scale=sr.std() / np.sqrt(sr.count()) )
def vp_sem(var):
    try: sr = pd.Series(var)
    except: return np.nan
    return sr.std() / np.sqrt(sr.count())
```

```
# Visual Python: ANOVA
# One-way ANOVA
vp_df = trainDF.dropna().copy()

from IPython.display import display, Markdown
_df = pd.DataFrame()
for k, v in  dict(list(vp_df.groupby('Pclass')['Survived'])).items():
    _df_t = v.reset_index(drop=True)
    _df_t.name = k
    _df = pd.concat([_df, _df_t], axis=1)

# Statistics
display(Markdown('### Statistics'))
display(pd.DataFrame(data={'Count':_df.count(),'Mean':_df.mean(numeric_only=True),'Std. Deviation':_df.std(numeric_only=True),'Min':_df.min(),'Max':_df.max(),
                           'Std. Error Mean':_df.apply(vp_sem),'Confidence interval':0.95,
                           'Lower':_df.apply(vp_confidence_interval).T[0],'Upper':_df.apply(vp_confidence_interval).T[1] }))

# Boxplot
import seaborn as sns
import warnings
with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=Warning)
    sns.boxplot(data=_df)
    plt.show()

# Equal Variance test (Levene)
from scipy import stats
_lst = []
_df.apply(lambda x: _lst.append(x.dropna()))
_res = stats.levene(*_lst, center='mean')
display(Markdown('### Equal Variance test (Levene)'))
display(pd.DataFrame(data={'Statistic':_res.statistic,'p-value':_res.pvalue}, index=['Equal Variance test (Levene)']))

# One-way ANOVA
import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm
_model  = smf.ols('Survived ~ C(Pclass)', vp_df)
_result = _model.fit()
_dfr = anova_lm(_result)
_dfr.loc['Total','df'] = _dfr['df'].sum()
_dfr.loc['Total','sum_sq'] = _dfr['sum_sq'].sum()
display(Markdown('### One-way ANOVA'))
display(_dfr)

# Post-hoc: Tukey HSD
from statsmodels.sandbox.stats.multicomp import MultiComparison
_res = MultiComparison(vp_df['Survived'], vp_df['Pclass']).tukeyhsd(alpha=0.05)
display(Markdown('### Post-hoc: Tukey HSD'))
display(_res.summary())
```

[ANOVA 결과 해석 방법]
* p-value < 0.05 : "ANOVA 테스트 결과, (Pclass에 따라 Survived에) 유의미한 영향을 미침"
* p-value > 0.05 : "ANOVA 테스트 결과, (Pclass에 따라 Survived에) 유의미한 영향을 미치지 않음"
<br>
→ 해당 분석 결과는 PR(>F)가 0.296023 이라는 결과가 나왔으므로 Pclass와 Survived 간의 통계적으로 유의미한 관계를 발견하지 못했습니다.

## Statistics > Regression

### Multiple linear regression (다중 선형 회귀 분석)

회귀 분석은 연속형 변수를 예측하는 데 사용됩니다. 종속 변수는 수치형이며, 예를 들어 주택 가격, 온도 등을 예측하는 데 사용됩니다.

다중 선형 회귀 분석의 경우에는 다음과 같은 총 4가지 method를 선택하여 분석을 진행할 수 있습니다.
- Enter : 모든 독립 변수를 동시에 모델에 포함하여 분석하는 방법입니다.
- Stepwise : 단계적 선택법(Stepwise Selection)으로, 전진 선택법과 후진 제거법을 혼합한 방법입니다.
- Backward : 후진 제거법(Backward Elimination)으로, 모든 변수를 포함한 상태에서 시작하여 불필요한 변수를 하나씩 제거해가는 방법입니다.
- Forward : 전진 선택법(Forward Selection): 아무 변수도 포함하지 않은 상태에서 시작하여 가장 유의미한 변수를 순차적으로 추가해가는 방법입니다. 유의미한 변수를 추가하고 불필요한 변수를 제거하는 과정을 반복합니다.

#### Method: Stepwise

```
# Visual Python: Regression
def vp_stepwise_select(df_x, df_y, alpha=0.05):
    select_list = list()
    while len(df_x.columns) > 0:
        col_list = list(set(df_x.columns)-set(select_list))
        sr_pval = pd.Series(index=col_list)
        for col in col_list:
            result = sm.OLS(df_y, sm.add_constant(df_x[select_list+[col]])).fit()
            sr_pval[col] = result.pvalues[col]
        best_pval = sr_pval.min()
        if best_pval < alpha:
            select_list.append(sr_pval.idxmin())
            while len(select_list) > 0:
                result = sm.OLS(df_y, sm.add_constant(df_x[select_list])).fit()
                sr_pval2 = result.pvalues.iloc[1:]
                worst_pval = sr_pval2.max()
                if worst_pval > alpha:
                    select_list.remove(sr_pval2.idxmax())
                else:
                    break
        else:
            break
    return select_list
```

```
# Visual Python: Regression
# Multiple linear regression > Method: Stepwise
vp_df = reg_tainDF.dropna().copy()

_selected_stepwise = vp_stepwise_select(vp_df[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked_label']], vp_df['Survived'])

# Model 1 - Dependent variable ~ Independent variable
from IPython.display import display, Markdown
import statsmodels.api as sm
_model  = sm.OLS(vp_df['Survived'], sm.add_constant(vp_df[_selected_stepwise[0]]))
_result = _model.fit()
display(Markdown('### Model 1 - Dependent variable ~ Independent variable'))
print(_result.summary())

# Model 1 - Multi-collinearity statistics
from statsmodels.stats.outliers_influence import variance_inflation_factor
_dfr = pd.DataFrame(_result.summary().tables[1].data[1:],columns=_result.summary().tables[1].data[0]).set_index('')
for i, col in enumerate(_model.exog_names[1:]):
    _vif = variance_inflation_factor(_model.exog, i+1)
    _dfr.loc[col,'Tolerance'] = 1/_vif
    _dfr.loc[col,'VIF'] = _vif
display(_dfr)

# Model 2 - Dependent variable ~ Stepwised variable
import statsmodels.api as sm
_model  = sm.OLS(vp_df['Survived'], sm.add_constant(vp_df[_selected_stepwise]))
_result = _model.fit()
display(Markdown('### Model 2 - Dependent variable ~ Stepwised variable'))
print(_result.summary())

# Model 2 - Multi-collinearity statistics
from statsmodels.stats.outliers_influence import variance_inflation_factor
_dfr = pd.DataFrame(_result.summary().tables[1].data[1:],columns=_result.summary().tables[1].data[0]).set_index('')
for i, col in enumerate(_model.exog_names[1:]):
    _vif = variance_inflation_factor(_model.exog, i+1)
    _dfr.loc[col,'Tolerance'] = 1/_vif
    _dfr.loc[col,'VIF'] = _vif
display(_dfr)

# Residual - Model 2
from IPython.display import display, Markdown
from scipy import stats
import statsmodels.api as sm
_predict =  _result.predict(sm.add_constant(vp_df[_model.exog_names[1:]]))
_residual = _result.resid
vp_residual = pd.DataFrame({'predict':_predict,'residual':_residual,
                            'predict_z':stats.zscore(_predict),'residual_z':stats.zscore(_residual)})
display(Markdown('### Residual - Model 2'))
display(vp_residual)

# Resisual Normality test (Shapiro-Wilk)
_res = stats.shapiro(vp_residual['residual_z'])
display(Markdown('### Residual Normality test (Shapiro-Wilk)'))
display(pd.DataFrame(data={'Statistic':_res.statistic,'p-value':_res.pvalue},index=['Resisual Normality test (Shapiro-Wilk)']))

import seaborn as sns
import warnings
with warnings.catch_warnings():

    # Residual histogram
    plt.subplot(2,2,1)
    warnings.simplefilter(action='ignore', category=Warning)
    sns.histplot(data=vp_residual, x='residual_z', kde=True)
    plt.title(f'Dependent variable: {_model.endog_names}')
    plt.xlabel('Regression Standardized residual')

    # Residual scatterplot
    plt.subplot(2,2,2)
    sns.scatterplot(data=vp_residual, x='predict_z', y='residual_z')
    plt.title(f'Dependent variable: {_model.endog_names}')
    plt.xlabel('Regression Standardized predicted value')
    plt.ylabel('Regression Standardized residual')

    plt.tight_layout()
    plt.show()
```

## Statistics > Correlation Analysis



Correlation Analysis(상관 분석)은 데이터 세트에서 변수 간의 관계를 이해하고 파악하는 데 사용됩니다.

```
# Visual Python: Correlation Analysis
def vp_confidence_interval_corr(x, y, method='pearson', alpha=0.05):
    try: x=pd.Series(x); y=pd.Series(y)
    except: return np.nan

    corr_func = {'pearson':stats.pearsonr,'spearman':stats.spearmanr,'kendall':stats.kendalltau}
    se_diff   = {'pearson':3,'spearman':3,'kendall':4}
    se_func   = {'pearson': lambda corr: 1,
                 'spearman':lambda corr: 1 + corr ** 2 / 2.,
                 'kendall': lambda corr: .437 }

    corr, pvalue = corr_func[method](x,y)

    z  = np.log((1 + corr) / (1 - corr)) / 2
    se = np.sqrt(se_func[method](corr) / (x.size - se_diff[method]))

    z_lower = z - stats.norm.ppf(1 - alpha / 2.) * se
    z_upper = z + stats.norm.ppf(1 - alpha / 2.) * se

    corr_lower = (np.exp(2 * z_lower) - 1) / (np.exp(2 * z_lower) + 1)
    corr_upper = (np.exp(2 * z_upper) - 1) / (np.exp(2 * z_upper) + 1)

    return corr, pvalue, corr_lower, corr_upper
```

```
# Visual Python: Correlation Analysis
vp_df = trainDF[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']].dropna().copy()

# Correlation Analysis
from scipy import stats
from IPython.display import display, Markdown
_dfr = pd.DataFrame()
for i, col1 in enumerate(vp_df.columns):
    for j, col2 in enumerate(vp_df.columns):
        if i >= j: continue
        if pd.api.types.is_numeric_dtype(vp_df[col1]) and pd.api.types.is_numeric_dtype(vp_df[col2]):
            _res = vp_confidence_interval_corr(vp_df[col1], vp_df[col2], method='pearson')
            _df_t = pd.DataFrame(data={'Variable1':col1,'Variable2':col2,'N':vp_df[col1].size,'Correlation coefficient':_res[0],
                                   'p-value':_res[1],'Lower(95%)':_res[2],'Upper(95%)':_res[3]}, index=[0])
            _dfr = pd.concat([_dfr, _df_t]).reset_index(drop=True)
display(Markdown('### Correlation Analysis: Pearson'))
display(_dfr)

# Correlation matrix: Pearson
from IPython.display import display
display(vp_df.corr(method='pearson', numeric_only=True).round(2))

# Chart
import seaborn as sns
import warnings
with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=Warning)

    # Heatmap
    sns.heatmap(vp_df.corr(method='pearson', numeric_only=True), annot=True, fmt='.2f', cmap='coolwarm')
    plt.title('Correlation heatmap: Pearson')
    plt.show()

    # Scatter matrix
    pd.plotting.scatter_matrix(vp_df)
    plt.show()
```

## Statistics > Nomality test


정규성 검정은 주로 다음과 같은 경우에 사용됩니다:

1. 통계적 가설 검정:
대표적인 예로 t-test, ANOVA, 회귀 분석 등의 통계적 가설 검정을 수행할 때, 데이터가 정규 분포를 따르는지 확인합니다. 이는 해당 분석 결과의 신뢰성을 높일 수 있습니다.
2. 통계적 모델링:
선형 회귀, 로지스틱 회귀, 시계열 분석 등의 통계적 모델링을 수행할 때, 종속 변수 및 독립 변수들이 정규 분포를 따르는지 확인합니다. 모델의 성능을 평가하거나 예측력을 향상시키는 데 도움이 됩니다.
3. 이상치 탐지:
데이터가 정규 분포를 따르지 않는 경우에는 이상치가 존재할 가능성이 높습니다. 따라서 정규성 검정을 통해 데이터의 분포를 확인하고 이상치를 탐지할 수 있습니다.
4. 통계적 요약 및 시각화:
데이터의 분포를 시각화하거나 통계적 요약을 할 때, 정규성 검정 결과를 함께 보여주어 데이터의 특성을 파악할 수 있습니다.

- Shapiro-Wilk test 정규성 검정 방법을 선택하였습니다.
- Age 변수에 대하여 정규성 검정을 실시하였습니다.

## Statistics > Equal Var. test

- Bartlett test
- Levene test
- Fligner test <br>

위의 등분산성 검정 방법 세가지 중 하나를 선택하여 검정할 수 있습니다.

1. Levene 검정: Levene 검정은 Bartlett 검정과 Fligner-Killeen 검정에 비해 이상치에 민감하지 않고, 비모수적인 방법에 가깝습니다. 이 방법은 그룹의 중앙값이 아닌 중위값을 사용하여 등분산성을 평가합니다. 이에 따라 이상치가 존재하는 경우에도 상대적으로 덜 영향을 받습니다.
Levene 검정은 등분산성을 평가할 때 일반적으로 많이 사용되는 방법 중 하나이며, 특히 비모수적인 방법을 선호하는 경우에 자주 활용됩니다.
2. Bartlett 검정: Bartlett 검정은 그룹의 데이터가 정규 분포를 따를 때 사용하기 적합한 방법입니다. 데이터가 정규성을 만족하지 않는 경우에는 Bartlett 검정의 결과가 신뢰할 수 없을 수 있습니다. 따라서 데이터의 정규성을 확인한 후에 Bartlett 검정을 사용하는 것이 좋습니다.
Bartlett 검정은 등분산성을 평가할 때 정규성 가정을 따르는 경우에 주로 사용되며, 정규성을 만족하는 데이터에 적용될 때 상대적으로 정확한 결과를 얻을 수 있습니다.
3. Fligner-Killeen 검정: Fligner-Killeen 검정은 비모수적인 방법으로, 데이터가 정규 분포를 따르지 않는 경우에 사용될 수 있습니다. 이 방법은 중위값 대신에 더 간단한 방법으로 등분산성을 평가합니다.
Fligner-Killeen 검정은 Bartlett 검정과 비교하여 이상치에 민감하지 않고, 비모수적인 방법을 선호하는 경우에 사용될 수 있습니다.

예제) 다음은 성별(Sex)에 따른 나이(Age) 변수를 가지고 등분산성을 평가하는 경우를 예를 들어 검정한 결과입니다.

```
# Visual Python: Equal Var. test
# Levene test
vp_df = trainDF.dropna().copy()
_df = pd.DataFrame()
for k, v in  dict(list(vp_df.groupby('Sex')['Age'])).items():
    _df_t = v.reset_index(drop=True)
    _df_t.name = k
    _df = pd.concat([_df, _df_t], axis=1)

# Variance
from IPython.display import display, Markdown
from scipy import stats
_dfr = _df.var().to_frame()
_dfr.columns = ['Variance']
display(_dfr)

# Levene test
_lst = []
_df.apply(lambda x: _lst.append(x.dropna()))
_res = stats.levene(*_lst, center='mean')
display(Markdown('### Levene test'))
display(pd.DataFrame(data={'Statistic':_res.statistic,'p-value':_res.pvalue},
                     index=['Equal Variance test (Levene)']))

# Histogram
import seaborn as sns
import warnings
with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=Warning)
    sns.histplot(_df, stat='density', kde=True)
    plt.title('Histogram')
    plt.show()
```

Levene test에서의 p-value 결과값이 0.310311로 0.05보다 크므로 즉, 등분산성을 가정할 수 있는 상황입니다. "성별에 따른 나이 변수의 등분산성 가정을 만족한다"라고 해석할 수 있습니다. 즉, 남성과 여성 그룹 간에 나이의 분산이 크게 다르지 않다는 것을 의미합니다.

## Statistics > Chi-square test

Chi-square 검정(Chi-square test)는 범주형 데이터 간의 관련성을 평가하기 위해 사용됩니다. 주로 두 개 이상의 범주형 변수가 서로 관련이 있는지를 확인하거나, 관측값과 기대값 간의 차이를 검정하는 데 사용됩니다.

예제)
- 귀무 가설(H0): 성별과 생존 여부는 독립적이다. (변수 간에 관련이 없다)
- 대안 가설(H1): 성별과 생존 여부는 종속적이다. (변수 간에 관련이 있다)

```
# Visual Python: Chi-square test
vp_df = trainDF.dropna().copy()

# Count plot
import seaborn as sns
import warnings
with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=Warning)
    sns.countplot(data=vp_df, x='Sex', hue='Survived')
    plt.show()

# Chi-square test of independence
from IPython.display import display, Markdown
from scipy import stats
_obs  = pd.crosstab(index=vp_df['Sex'], columns=vp_df['Survived'])
_res1 = stats.chi2_contingency(_obs)
_res2 = stats.chi2_contingency(_obs, lambda_='log-likelihood')

# Cross tabulation: Count
_dfc = pd.crosstab(index=vp_df['Sex'],columns=vp_df['Survived'],margins=True,margins_name='Total')
_dfc = _dfc.reset_index().reset_index()
_dfc[' '] = 'Count'

# Cross tabulation: Expected count
_dfe = pd.DataFrame(_res1.expected_freq, index=_obs.index, columns=_obs.columns).round(1)
_dfe.loc['Total',:] = _dfe.sum(axis=0)
_dfe.loc[:,'Total'] = _dfe.sum(axis=1)
_dfe = _dfe.reset_index().reset_index()
_dfe[' '] = 'Expected count'

# Cross tabulation: Count + Expected count
display(Markdown('### Cross tabulation'))
display(pd.concat([_dfc, _dfe]).set_index(['Sex',' ']).sort_values('index').drop('index',axis=1))

# Chi-square test
display(Markdown('### Chi-square test'))
display(pd.DataFrame(data = {'Value':[_res1.statistic,_res2.statistic,vp_df.dropna().shape[0]],
                             'df':[_res1.dof,_res2.dof,np.nan],
                             'p-value(two-sided)':[_res1.pvalue,_res2.pvalue,np.nan]},
                     index= ['Pearson Chi-square','Likelihood ratio','N of valid cases']))

# Cramers' V coefficient
_X2 = stats.chi2_contingency(_obs)[0]
_sum = _obs.sum().sum()
_minDim = min(_obs.shape)-1
display(Markdown('### Cramers V coefficient'))
display(pd.DataFrame(data={'Value':np.sqrt((_X2/_sum) / _minDim)}, index=['Cramers V coefficient']))
```

Pearson Chi-square의 p-value값이 0.05보다 작으므로 귀무가설을 기각하여 두 변수(성별, 생존)는 통계적으로 유의한 관계가 있습니다. 따라서, 성별이 생존 여부에 영향을 준다고 결론지을 수 있습니다.

## Statistics > Student's t-test


One-sample t-test, Independent two-sample t-test, Paired-samples t-test → 3가지 중 하나를 선택하여 검정을 진행할 수 있습니다.

1. One-sample t-test (일표본 t-검정)
- 상황: 하나의 표본 집단에서 모집단의 평균을 비교하는 경우에 사용됩니다.
- 예시: 어떤 학급의 학생들의 평균 시험 점수가 전체 학생들의 평균과 같은지 비교하는 경우.
- 가설: 연구자가 표본의 평균이 어떤 특정 값과 같은지 여부를 검정합니다.
2. Independent two-sample t-test (독립표본 t-검정)

- 상황: 두 개의 독립된 표본 집단 간의 평균 차이를 비교하는 경우에 사용됩니다.
- 예시: 남성 그룹과 여성 그룹의 평균 키가 같은지 비교하는 경우.
- 가설: 두 그룹의 평균이 같은지 여부를 검정합니다.
3. Paired-samples t-test (대응표본 t-검정)

- 상황: 한 집단의 관측값들이 쌍을 이루는 경우, 같은 사람 또는 동일한 대상에 대해 조사한 경우에 사용됩니다.
- 예시: 동일한 표본 집단에서 전·후 시험 점수의 차이가 있는지 비교하는 경우.
- 가설: 짝을 이루는 관측값들 간의 평균 차이가 있는지 여부를 검정합니다.

예제) Independent two-sample t-test (독립표본 t-검정)
*   귀무 가설(H0): 성별에 따른 생존 여부의 평균은 같다.
*   대안 가설(H1): 성별에 따른 생존 여부의 평균은 다르다.



```
# Visual Python: Student's t-test
# Independent two-sample t-test
vp_df1 = trainDF[(trainDF['Sex'] == 'male')]['Survived'].dropna().copy()
vp_df2 = trainDF[(trainDF['Sex'] == 'female')]['Survived'].dropna().copy()

# Normality test (Shapiro-Wilk)
from IPython.display import display, Markdown
from scipy import stats
_res1 = stats.shapiro(vp_df1)
_res2 = stats.shapiro(vp_df2)
display(Markdown('### Normality test (Shapiro-Wilk)'))
display(pd.DataFrame(data={'Statistic':[_res1.statistic,_res2.statistic],'p-value':[_res1.pvalue,_res2.pvalue]},
                    index=[['Normality test (Shapiro-Wilk)' for i in range(2)],['Variable1','Variable2']]))

# Equal Variance test (Levene)
display(Markdown('### Equal Variance test (Levene)'))
_res = stats.levene(vp_df1, vp_df2, center='mean')
display(pd.DataFrame(data={'Statistic':_res.statistic,'p-value':_res.pvalue}, index=['Equal Variance test (Levene)']))

# Statistics
display(Markdown('### Statistics'))
display(pd.DataFrame(data={'N':[vp_df1.size,vp_df2.size],
                           'Mean':[vp_df1.mean(),vp_df2.mean()],
                           'Std. Deviation':[vp_df1.std(),vp_df2.std()],
                           'Std. Error mean':[vp_df1.std()/np.sqrt(vp_df1.size),
                                              vp_df2.std()/np.sqrt(vp_df2.size )]},
                     index=[['Statistics' for i in range(2)],['Variable1','Variable2']]))

# Independent two-sample t-test
_res1 = stats.ttest_ind(vp_df1, vp_df2, equal_var=True,  alternative='two-sided')
_res2 = stats.ttest_ind(vp_df1, vp_df2, equal_var=False, alternative='two-sided')
display(Markdown('### Independent two-sample t-test'))
display(pd.DataFrame(data={'Statistic':[_res1.statistic,_res2.statistic],'Alternative':['two-sided' for i in range(2)],
                           'p-value':[_res1.pvalue,_res2.pvalue],
                           'Mean difference':[vp_df1.mean()-vp_df2.mean() for i in range(2)]},
                     index=[['Independent two-sample t-test' for i in range(2)],['Equal variance' for i in range(2)],[True,False]]))
display(Markdown('If equal_var is False, perform Welch\'s t-test, which does not assume equal population variance'))
```

p-value값이 2.283601e-61가 나왔으므로 귀무가설을 기각하여 즉, 성별에 따라 생존 여부에 대한 평균적인 차이가 있다고 할 수 있습니다.
